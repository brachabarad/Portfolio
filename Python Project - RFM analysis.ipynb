```python
# Import the relevant libraries
import numpy as np
import plotly.express as px
import pandas as pd
```

```python
# Load dataset
cor_df = pd.read_csv('/content/drive/MyDrive/datasets/olist_customer_order_revenue.csv')

# Print DataFrame shape
print(cor_df.shape)

# Preview DataFrame
cor_df.head(10)
```

```python
# Create a dictionary of column names and aggregation operations
to_agg={'order_id':'count',
             'sales_amount':'sum'}

# Group by customer ID and aggregate data
fm_df = cor_df.groupby('customer_unique_id').agg(to_agg)

```

If done correctly, your new DataFrame `fm_df` should have `96090` rows and `2` columns. Confirm by printing the DataFrame's `shape`. Preview the first ten rows as well.

```python
# Print DataFrame shape

print(fm_df.shape)

# Preview DataFrame
fm_df.head(10)

```

```python
# Load the customer recency data 
recency = pd.read_csv('/content/drive/MyDrive/datasets/customer_recency.csv')

# Preview the customer regency data
recency.head()
```

```python
# Join fm_df with the recency DataFrame
rfm_df = pd.merge(fm_df, recency, left_on='customer_unique_id', right_on='customer_unique_id', how='inner')

# Preview joined data
rfm_df.head()
```

```python
# Rename columns
rfm_df.rename(columns ={'order_id':'frequency','sales_amount':'monetary'}, inplace=True)

# Preview data
rfm_df.head(10)
```

```python
# Create a histogram of frequency 

px.histogram(rfm_df, x='frequency', title='Frequency of Purchases')
```

```python
def frequency_grade(frequency): 
    '''
    Given a frequency value, function compares the value to 
    the frequency grading scale and returns the appropriate
    grade. 
    '''
    # check if frequency is equal to 1
    if frequency==1:
        
        # if so, assign letter grade D
        grade='D'
        
    # check if frequency is equal to 2  
    elif frequency==2:
        
        # if so, assign letter grade C
        grade='C'
        
    # check if frequency is less than or equal to 4 
    elif frequency<=4:
        
        # if so, assign letter grade B
        grade='B'
        
    # otherwise, assign letter grade A
    else:
        grade='A'
        
    # return letter grade
    return grade
```

```python

# apply frequncy_grade to rfm_df 
rfm_df['frequency_grade'] = rfm_df['frequency'].apply(frequency_grade)

# preview DataFrame
rfm_df.head()
```

```python
# Create a bar chart of total customer spending by grade
px.histogram(rfm_df, x='frequency_grade',
             labels={'frequency_grade':'Frequency Grade', 'y':'Number of Customers'},
             title = 'Frequency Grades of Olist Customers',
             category_orders = {'frequency_grade': ['A', 'B', 'C', 'D']})
```

```python
# Create a histogram of recency
px.histogram(rfm_df, x='recency', nbins=50, title='Recency of Purchases')
```
